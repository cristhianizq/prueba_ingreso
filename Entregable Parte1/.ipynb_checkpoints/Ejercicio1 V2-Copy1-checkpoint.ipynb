## Funciones

```python
def conteoXGrupo(df, group_field):
    grouped = df.groupby(group_field, as_index=False).size()
    #grouped = df.groupby(group_field).size()
    grouped = pd.DataFrame(grouped).reset_index()    
    grouped=grouped.set_index('index')
    grouped=grouped.sort_values(by='size', ascending = False)    
    grouped.columns = [group_field, 'conteo']
    grouped["participacionRelativa"] = round(grouped['conteo']/grouped['conteo'].sum()*100,2)
    grouped["participacionAcumulada"] = round(grouped['conteo'].cumsum()/grouped['conteo'].sum()*100,2)
    return grouped
```

```python
#función para gráficar
def graficaPareto(df, group_field,frec):
    fig, ax = plt.subplots(figsize=(50,20))
    # Plot bars (i.e. frequencies)
    ax.bar(df[group_field], df[frec])
    ax.set_title("Ocurrencia por "+group_field, fontsize=60)
    ax.set_xlabel(group_field,fontsize=36)
    plt.xticks(fontsize=36)
    plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
    ax.set_ylabel("Frecuencia",fontsize=40);
    plt.yticks(fontsize=40)
    # Second y axis (i.e. cumulative percentage)
    ax2 = ax.twinx()
    ax2.plot(df[group_field], df["participacionAcumulada"], color="red", marker="D", ms=7)
    ax2.axhline(80, color="black", linestyle="dashed")
    ax2.yaxis.set_major_formatter(PercentFormatter())
    plt.yticks(fontsize=40)
    ax2.set_ylabel("Participacion Acumulada",fontsize=40);
```

```python
dataManejoFechas['fechaCreacionAjust'] = pd.to_datetime(dataManejoFechas['Created Date'])
dataManejoFechas['fechaCreacion'] = [d.date() for d in dataManejoFechas['fechaCreacionAjust']]
dataManejoFechas['horaCreacion'] = [d.time() for d in dataManejoFechas['fechaCreacionAjust']]
dataManejoFechas['AñoCreacion']=pd.DatetimeIndex(dataManejoFechas['fechaCreacion']).year.astype(int)
dataManejoFechas['MesCreacion']=pd.DatetimeIndex(dataManejoFechas['fechaCreacion']).month.astype(int)
dataManejoFechas['PeriodoCreacion']=pd.to_datetime((dataManejoFechas['AñoCreacion']*10000+dataManejoFechas['MesCreacion']*100+1).apply(str),format='%Y%m%d')
```
